##Типизация
Присвойте соответствующим колонкам один из типов (category, float, int)

```python
##df = df.astype({'age': np.int8, 'Medu': np.float64})

df["age"] = df['age'].astype('int')
df["Medu"] = df['Medu'].astype('float')
df["school"] = df['school'].astype('string')
df["sex"] = df['sex'].astype('category')

#Вывести типы колонок тут
df.dtypes[['age','Medu','school','sex']]
##df.info()
```
